#### 2- Manuellement

```python
df['school']=df['school'].map({'GP':0, 'MS':1})
df['sex']=df['sex'].map({'M':0 ,'F':1})
df['address']=df['address'].map({'R':0 ,'U':1})
df['famsize']=df['famsize'].map({'LE3':0 ,'GT3':1})
df['Pstatus']=df['Pstatus'].map({'A':0 ,'T':1})
df['Mjob']=df['Mjob'].map({'at_home':0 ,'services':1, 'teacher':2, 'health':3, 'other':4})
df['Fjob']=df['Fjob'].map({'at_home':0 ,'services':1, 'teacher':2, 'health':3, 'other':4})
df['famsup']=df['famsup'].map({'no':0, 'yes':1})
df['reason']=df['reason'].map({'course':0 ,'home':1, 'reputation':2, 'other':3})
df['guardian']=df['guardian'].map({'mother':0 ,'father':1, 'other':2})
df['schoolsup']=df['schoolsup'].map({'no':0, 'yes':1})
df['paid']=df['paid'].map({'no':0, 'yes':1})
df['activities']=df['activities'].map({'no':0, 'yes':1})
df['nursery']=df['nursery'].map({'no':0, 'yes':1})
df['higher']=df['higher'].map({'no':0, 'yes':1})
df['internet']=df['internet'].map({'no':0, 'yes':1})
df['romantic']=df['romantic'].map({'no':0, 'yes':1})
```

```python
## affichage de notre dataset après encodage
df.transpose()
```

# Selection

On peut voir les features qui affectent nos targets

#### 1) Selection pour G1

```python
df_features1 = df.drop(['G1','G2','G3'], axis=1)

#notre target est G1
target1 = df['G1']
test1 = SelectKBest(score_func=chi2, k=10)
fit1 = test1.fit(df_features1, target1)
#choix des features
feat_importances1 = pd.Series(fit1.scores_, index=df_features1.columns)
topFeatures1 = feat_importances1.nlargest(50).copy().index.values
```

```python
topFeatures1
```

#### 2) Selection pour G3

```python
df_features2 = df.drop(['G2','G3'], axis=1)
## notre target est G3
target2 = df['G3']
test2 = SelectKBest(score_func=chi2, k=10)
fit2 = test2.fit(df_features2, target2)
## choix des features
feat_importances2 = pd.Series(fit2.scores_, index=df_features2.columns)
topFeatures2 = feat_importances2.nlargest(50).copy().index.values
```
